# Relativistic and Non-Relativistic Specis in Cosmology:

The exact expression for the number density of a fermion of mass m in thermal equilibrium
with zero chemical potential,  



\begin{align}
n& =  g\int_{0}^{\infty} \frac{d^3 p}{(2\pi)^3}\frac{1}{e^\frac{E}{T}+1}\ \tag{1}\\ 
&n = g \int_{0}^{\infty} \frac{d^3 p}{(2\pi)^3}\frac{1}{e^\frac{\sqrt{p^2+m^2}}{T}+1}\tag{2}
\end{align}


For Isotropic distribution $d^3 p=4\pi p^2 dp $ equation (2) can be written as following,

\begin{align}
n &= g \int_{0}^{\infty} \frac{4\pi p^2 dp}{(2\pi)^3}\frac{1}{e^\frac{E}{T}+1}\ \tag{3}\\ 
&n= \frac{2g}{(2\pi)^2}\int_{0}^{\infty} \frac{p^2}{e^\frac{\sqrt{p^2+m^2}}{T}+1} dp\tag{4}
\end{align}
Using $x=\frac{p}{T}$ and $y=\frac{m}{T}$ , and $dx=\frac{dp}{T}$, equation (4) becomes,

\begin{align}
&n= \frac{2g}{(2\pi)^2}\int_{0}^{\infty} \frac{x^2}{e^\frac{\sqrt{T^2x^2+T^2 y^2}}{T}+1} dx\tag{5}\\
&\frac{n(y)}{T^3}= \frac{g}{2\pi^2}\int_{0}^{\infty} \frac{x^2}{e^{\sqrt{x^2+y^2}}+1} dx\tag{6}\\
\end{align}

```python
y=np.linspace(0.1,10,100)  #y=m/T=100 is taken

res=[]
for i in np.arange(100): #for i in np.arange(y.size)
    integrand=lambda x: (x**2/np.pi**2)*(1/np.exp(np.sqrt(x**2+y[i]**2)))  #equation (6)
    integral,error=quad(integrand,0,np.inf)
    res.append(integral)
```

```python
n_per_T_cube=np.array(res)  #n(y)/T^3
```

Number density for Relativistic fermions: $$n_R=g\frac{3}{4}\frac{\zeta{(3)}}{\pi^2}T^3$$

$$\frac{n}{T^3}=g\frac{3}{4}\frac{\zeta{(3)}}{\pi^2} \tag{7}$$

Number density for Non-Relativistic Fermions: $$n_{NR} = g(\frac{mT}{2\pi})^{3/2}e^{-\frac{(m-\mu)}{T}}$$

$$\frac{n_{NR}{(y=\frac{m}{T}})}{{T^3}} = g(\frac{y}{2\pi})^{3/2}e^{-\frac{(m-\mu)}{T}}\tag{8}$$


```python
g=2  #electrons have two spin states (up, down)

#m=0.511 MeV

def n_r(y): #relativistic case: number density of electrons
   
    return ((3*g)/(4*np.pi**2))*zeta(3)

def n_nr(y): #non-relativistic case: number density of electrons
    
    return (g*(y/(2*np.pi))**(3/2))*np.exp(-y)  #mu=0 
```

```python
n_r=n_r(y)
```

```python
n_nr=n_nr(y)
```

```python
plt.loglog(y,n_per_T_cube,"go-",label="Rel+Non-Rel", markersize="3")
plt.loglog(y,n_nr,"ro",label="Non Relativistic", markersize="3")
plt.axhline(y= n_r, xmin=0.01, xmax=10, color="blue", linestyle=":",lw=3.0, label="Relativistic")
plt.xlabel("y=m/T", fontsize=16)
plt.ylabel(r"$n(y)/T^3$",fontsize=16)
plt.legend(loc="best", edgecolor="black")
```

Both Relativistic $(y<<1)$ and Non-Relativistic $(y>>1)$ limit for
the number density of the fermions. It is evident that the result using eq(6) match the exact
result plotted with eq(7) and eq(8) in the appropriate limit.

For Massive neutrinos:
\begin{align}
 n_{\nu}& =  g\int_{0}^{\infty} \frac{d^3 p}{(2\pi)^3}\frac{1}{e^\frac{E}{T_{\nu}}+1}\ \tag{10}\\ 
 &\rho_{\nu} = g \int_{0}^{\infty} \frac{d^3 p}{(2\pi)^3}\frac{E_{\nu}}{e^\frac{p}{T_{\nu}}+1}\tag{11}\\
&\rho_{\nu}= g \int_{0}^{\infty} \frac{d^3 p}{(2\pi)^3}\frac{\sqrt{p^2+m_{\nu}^2}}{e^\frac{p}{T_{\nu}}+1}\tag{12}\\
&\rho_{\nu}=\frac{g}{2\pi^2}\int_{0}^{\infty} {p^2 dp}\frac{\sqrt{p^2+m_{\nu}^2}}{e^\frac{p}{T_{\nu}}+1}\tag{13}\\
\end{align}

De-Broglie $p=\frac{h}{\lambda}=\frac{2\pi\hbar}{\lambda} =\frac{2\pi(\hbar=1)}{\lambda}$, since $ \lambda\propto a$  we get, $p=\frac{2\pi}{a}=2\pi(1+z)$, and $T_{\nu}\approx\frac{1}{a}=1+z$, $x=\frac{p}{T_{\nu}}$

Taking one of the neutrinos massless and other two massive,  $m_1=0,m_2=0.01$ eV , $m_3=0.05$ eV then Equation(13) becomes,

$$ \rho_{\nu}(z)=\frac{gT_{\nu}^3}{2\pi^2}\int_{0}^{\infty} {x^2 dx }\frac{\sqrt{T_{\nu}^2x^2+m_{\nu}^2}}{e^x+1}\tag{14}$$

Equation(14) is for individual specis, for three neutrinos, $\rho_{\nu}$ takes the form,  
$$ \rho_{\nu,1}(z)=\frac{gT_{\nu}^4}{2\pi^2}\int_{0}^{\infty} {x^2dx }\frac{\sqrt{x^2+m_{1}^2(1+z)^2}}{e^{x}+1}\tag{15}$$
$$ \rho_{\nu,2}(z)=\frac{gT_{\nu}^4}{2\pi^2}\int_{0}^{\infty} {x^2dx }\frac{\sqrt{x^2+m_{2}^2(1+z)^2}}{e^{x}+1}\tag{16}$$
$$ \rho_{\nu,3}(z)=\frac{gT_{\nu}^4}{2\pi^2}\int_{0}^{\infty} {x^2dx }\frac{\sqrt{x^2+m_{3}^2(1+z)^2}}{e^{x}+1}\tag{17}$$

$$ \rho_{\nu}(z)=\rho_{\nu,1}(z)+\rho_{\nu,2}(z)+\rho_{\nu,3}(z) \tag{18}$$

Energy density of Photon $\rho_{\gamma}=\frac{\pi^2}{15} T_{\gamma}^4$ and $\frac{T_{\nu}}{T_{\gamma}}=(\frac{4}{11})^{1/3}$, putting all together $\frac{\rho_{\nu}(z)}{\rho_{\gamma}}$ have the expression,

$$\frac{\rho_{\nu}(z)}{\rho_{\gamma}}=g\frac{15}{2\pi^4}(\frac{4}{11})^{4/3}\int_{0}^{\infty} \frac{x^2dx}{e^{x}+1} ({\sqrt{x^2+m_{1}^2(1+z)^2}}+{\sqrt{x^2+m_{2}^2(1+z)^2}}+{\sqrt{x^2+m_{3}^2(1+z)^2}} \space)\tag{19}$$

```python
g=2 #neutrinos &antineutrinos

factor=((g*15)*(4/11)**(4/3))/(2*np.pi**4)

m_1=0    #lightest mass
m_2=0.01 #eV
m_3=0.05 #eV

z=np.linspace(0,1000,100)  

res=[]

for i in np.arange(100): #for i in np.arange(z.size)
    
    #equation(19)
    integrand=lambda x: factor*(x**2/(np.exp(x)+1))*np.sqrt((x**2+m_2**2*(1+z[i])**2)+(x**2+m_2**2*(1+z[i])**2)+(x**2+m_3**2*(1+z[i])**2))
    
    integral,error=quad(integrand,0,np.inf)
    
    res.append(integral)
    
Integration_result=np.asarray(res)  #np.array(res)
```

```python
fig, ax=plt.subplots()
ax.semilogx(z,Integration_result,"mo-", markersize="4")
ax.set_xlabel("Redshift z", fontsize=14)
ax.invert_xaxis()
ax.set_ylabel(r"$\frac{\rho_{\nu}(z)}{\rho_{\gamma}}$", fontsize="16")
```

The above plot shows ratio of energy density of neutrinos to the photons as function of redshift. Here redshift values upto 1000 is taken.

```python
#Making it a function and doing the same thing in a efficient way: so that we can play with function arguments(masses)

g=2 #neutrinos & antineutrinos
factor=((g*15)*(4/11)**(4/3))/(2*np.pi**4)

#m_1=0    #lightest mass
#m_2=0.01 #eV
#m_3=0.05 #eV

z=np.linspace(0,1000,100)  
res=[]

def ratio(m_1,m_2,m_3):  #ratio of energy density of all three neutrinos to the photon
    for i in np.arange(100): #for i in np.arange(z.size)
        integrand=lambda x: factor*(x**2/(np.exp(x)+1))*np.sqrt((x**2+m_2**2*(1+z[i])**2)+(x**2+m_2**2*(1+z[i])**2)+(x**2+m_3**2*(1+z[i])**2))
        integral,error=quad(integrand,0,np.inf)
        res.append(integral)
    Integration_result=np.asarray(res)
    return Integration_result
```
