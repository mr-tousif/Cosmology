# Big Bang Nucleosynthesis (BBN)


BBN is of particular importance, as it produced the light elements we see in
the universe and can be used to constrain cosmology. BBN happened when the temperature of the universe cooled to 1 MeV. At that point in time, the cosmic plasma consisted of: Relativistic particles in equilibrium: photons, electrons and positrons, Nonrelativistic particles: baryons(protons, neutrons).


Baryons to Photons Ratio: $\eta_b=\frac{n_b}{n_\gamma}=6\times 10^{-10}\left(\frac{\omega_b h^2}{0.022}\right)$


Number density for Non-Relativistic Fermions in plasma: $m>>T$ $$n_{NR} = g\left(\frac{mT}{2\pi}\right)^{3/2}e^{-\frac{(m-\mu)}{T}}\tag{1}$$


Neutrons and protons can interconvert via weak interactions in equilibrium: $n\leftrightarrow p+e^{-}+\bar{\nu}$, $p+e^{-}\leftrightarrow n+\nu$, $p+\bar{\nu}=n+e^{+}$

Number density of neutrons in the plasma, $n_n= g\left(\frac{m_n T}{2\pi}\right)^{3/2}e^{-\frac{(m_n-\mu_n)}{T}}\tag{2}$

Number density of protons in the plasma, $n_p= g(\frac{m_p T}{2\pi})^{3/2}e^{-\frac{(m_p-\mu_p)}{T}}\tag{3}$


Taking ratio of $n_n$ to $n_p$ at equilibrium ($\mu\approx 0$) we get, $$\frac{n_n}{n_p}=e^{\frac{-(m_n-m_p)}{T}}=e^{-\frac{Q}{T}}\tag{4}$$

The prefactor is close to unity since neutron mass $m_n=939.6$ MeV and proton mass $m_p=938.3$ MeV  is close to each other. $Q\equiv{m_nâˆ’m_p}=1.3$ MeV.

The time-temperature relation is $t=132 s \left(\frac{0.1 MeV}{T}\right)^2\tag{5}$. 
$$T={0.1 MeV}{\sqrt{\frac{132s}{t}}}\tag{6}$$

It is evident from eq(4)  as temperatues goes down the neutron fraction goes down.The main task of this section is to find what happens in the real world where weak interactions are not so efficient. It is convenient to define,

$$X_n=\frac{n_n}{n_n+n_p}\tag{7}$$ $X_n$ is the ratio to neutrons to total nuclei. Rewriting eq(7) & eq(4)  we have,

$$\frac{n_n}{n_p}=\frac{X_n}{1-X_n}=e^{-\frac{Q}{T}}\tag{8}$$

```python
#importing modules

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
m_n=939.6 #MeV
m_p=938.3 #MeV
t=np.linspace(1e-2,1,100) #seconds
Q=1.3 #MeV
T=np.linspace(5,1e-2,100)  #MeV

def n_p_ratio(T):
    res=np.exp(-Q/T)
    return res

def n_p_t(t):
    T=0.1*np.sqrt(132/t) #MeV
    #res=(m_n/m_p)**(3/2)*np.exp(-Q/T)
    res=np.exp(-Q/T)
    return res
```

```python
 n_p_ratio(T)
```

```python
fig,ax=plt.subplots()
ax.semilogx(T,n_p_ratio(T),"bo-")
ax.invert_xaxis()
ax.set_xlabel(r"T[MeV]")
ax.set_ylim(0,1)
ax.set_ylabel(r"$\frac{n_n}{n_p}$", fontsize=16)
```

Conversions between protons and neutrons ceased to be efficient at $T=0.7$ MeV . At this time, the ratio n/p was about $\frac{n_n}{n_p}=e^{-\frac{1.3}{0.7}}\approx\frac{1}{6}$.

Afterwards neutrons began to decay hence$n_n=n_{n0} e^{-\frac{t}{\tau_n}}\tag{9}$ with $n_{n0}=n_p e^{-\frac{Q}{T}}$ 
and eq(9) takes the form,$n_n=n_p e^{-\frac{Q}{T}} e^{-\frac{t}{\tau_n}}\tag{10}$

So the n/p ratio becomes, $$\frac{n_n}{n_p}=e^{-\frac{Q}{T}} e^{-\frac{t}{\tau_n}}\approx \frac{1}{6}e^{-\frac{t}{\tau_n}}\tag{11}$$

```python
t=np.linspace(1e-2,1e3,100)
tau_n=880 #seconds
def n_p_ratio_l(t):
    res=(1/6)*np.exp(-t/tau_n)
    return res
```

```python
fig,ax=plt.subplots()

twin1 = ax.twiny()
p1, =ax.semilogx(t,n_p_ratio_l(t),"ro-",label=r"Free neutron decay $e^{-\frac{Q}{T}}e^{-\frac{t}{\tau_n}}$")
p2, =ax.semilogx(t,n_p_t(t), "darkgreen",marker="o", label=r"Equilibrium n/p ratio $e^{-\frac{Q}{T}}$")
ax.set_ylim(0,1)
ax.set_xlim(1e-2,1000)
twin1.set_xlim(5, 0)
ax.set_xlabel("t [seconds]")
ax.set_ylabel(r"$\frac{n_n}{n_p}$", fontsize=16)
twin1.set_xlabel("T in MeV")
ax.legend(handles=[p1, p2])
ax.tick_params(axis="y",direction="in",pad=5)
ax.tick_params(axis="x",direction="in",pad=5)
twin1.tick_params(axis="x",direction="in",pad=3)
#ax.tick_params(axis="x",direction="in", pad=-15) #making ticks inside
#ax.grid()
plt.show()
```
