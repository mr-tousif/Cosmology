


since the dark matter density today is $\Omega_{DM}h^2=0.119$, and the contribution of neutrinos to the critical density is $$\Omega_{\nu}h^2=\frac{\sum{m_{\nu,i}}}{94 eV}$$

For Neutrinos to be dark matter, It has to be $\Omega_{DM}h^2=\Omega_{\nu}h^2$ so we get,

$$\frac{\sum{m_{\nu,i}}}{94\,eV}=0.119$$

$${\sum{m_{\nu,i}}}=11.186\, eV$$ so we can can conclude the neutrino to be a dark matter at individual mass has to be 1/3 of $\sum m_{\nu,i}$


Neutrinos are relativistic at early, they have a lot of pressure can not form gravitionally bound structure. One way to see this is to compute the free-streaming length $r_{fs}$ of neutrinos

\begin{equation}
r_{fs}=\int v\,dt=\int_{0}^{1} v\,\frac{da}{aH(a)}\tag{1}
\end{equation}

where $v=\frac{p}{E}$, neutrino speed. E is energy and p is momemtum $E=\sqrt{p^2+m^2}$. Taking $p\approx 3.0\, T_{\nu,0}$ and $T_{\nu,0}=\left(\frac{4}{11}\right)^{1/3}\,T_{\gamma,0}$ where $T_{\gamma,0}$ is photon temprature which is $2.348\times 10^{-4}$eV today.

$$H(a)=H_0\sqrt{\Omega_r\,a^{-4}+\Omega_m\,a^{-3}+\Omega_k\,a^{-2}+\Omega_\Lambda}$$

```python
import numpy as np
import scipy as sp
from scipy.integrate import quad
import matplotlib.pyplot as plt
```

```python
# Universe filled with matter, radiation, dark energy with zero curvature

Omega_m=0.3        #Matter density
Omega_r=9.1*1e-5      #Radiation density
Omega_k=0
Omega_Lambda=1-Omega_m-Omega_r-Omega_k  #cosmological constant(Dark Energy)

H_0=67.5   #Hubble Constant value today 


T_gamma=2.348*1e-4
T_v=((4/11)**(1/3))*T_gamma
m_v=11.186/3

def p(a):
    p=(3*T_v)/a
    return p

def E(a):
    E=np.sqrt(p(a)**2+m_v**2)
    return E

def v(a):
    v=p(a)/E(a)
    return v

def H(a):
    return H_0*np.sqrt(Omega_r*a**-4+Omega_m*a**-3+Omega_k*a**-2*Omega_Lambda)

c=3*10**5 #km/sec

integrand=lambda a:(v(a)*c)/(a*H(a))  # equation(1)

integral, integral_error=quad(integrand,0,1)  # Integration limit 0 to 1, as scale factor a(t)=1 today
```

```python
integral #Mpc
```

```python
integral*1e3 #kpc
```
