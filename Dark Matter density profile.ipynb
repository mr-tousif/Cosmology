```python
import numpy as np
import matplotlib.pyplot as plt
```

First, the density distribution appears to be approximately universal and well-modeled by the
Navarro-Frenk-White (NFW) profile given by;

\begin{align}
\rho_{NFW}=\frac{\rho_o}{\frac{r}{r_s}({1+\frac{r}{r_s}})^2}\tag{1}
\end{align}

\begin{align}
\rho_{Ein(r)}=\rho_0{exp\left[\frac{-2}{\gamma}\left(\left({\frac{r}{r_s}}\right)^\gamma-1\right)\right]}\tag{2}
\end{align}
where $r_s=20$ Kpc is the scale radius and $\gamma=0.17$ 

\begin{align}
\rho_{Burk(r)}=\frac{\rho_o}{\left(1+\frac{r}{r_s}\right)\left(1+\left(\frac{r}{r_s}\right)^2\right)}\tag{3}
\end{align}

```python
rho_o=0.3 #GeV/cm^3
r_s=20 #kpc
gamma=0.17
r=np.linspace(1e-2,1e2,100)

def rho_NFW(r):
    return rho_o/((r/r_s)*(1+(r/r_s)))**2

def rho_Ein(r):
    factor=(r/r_s)**gamma-1
    return rho_o*np.exp(-2*factor/gamma)

def rho_Burk(r):
    return rho_o/((1+(r/r_s))*(1+(r/r_s)**2))

```

```python
plt.loglog(r,rho_NFW(r),"--", label="NFW")
plt.loglog(r,rho_Ein(r),"--", label="Einasto")
plt.loglog(r,rho_Burk(r),"-", label="Burkert")
plt.legend(loc="best")
plt.xlabel(r'$r_s$ [kpc]')
plt.ylabel("Density $[GeV/cm^3]$")
```
